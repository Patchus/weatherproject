```python
weather = pd.read_csv('weather_daily.csv')
weather = weather[['city','date','high_f']].drop_duplicates()
print weather[0:50]
nycd = {}

cities = weather['city'].drop_duplicates()

for city in cities:
    nycd[city] = list(weather['high_f'][weather['city']==city])
    
index = range(1, 11, 1)
nycd['dates'] = index
```

```python
import vincent
vincent.core.initialize_notebook()
lines = vincent.Line(nycd, iter_idx='dates')
lines.legend(title='City')
lines.axis_titles(x='Dates', y='High')
```
